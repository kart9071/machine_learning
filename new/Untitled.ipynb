```python
import mglearn
import matplotlib.pyplot as plt
%matplotlib inline
X, y = mglearn.datasets.make_forge()
# plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
print(mglearn.datasets.make_forge())
```

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target")
print(mglearn.datasets.make_wave())
```

```python
from sklearn.datasets import make_classification

# Generate a synthetic dataset with 2 classes, 10 total features, 5 informative features, 3 redundant features, and 2 repeated features
X, y = make_classification(n_classes=2, n_features=10, n_informative=5, n_redundant=3, n_repeated=2,n_samples=10)

# Print the shape of X
print(X.shape)
print(X)
print(y)

# Plot the dataset
import matplotlib.pyplot as plt
plt.scatter(X[:, 5], X[:, 6], c=y)
```

```python
X,y=mglearn.datasets.make_wave(n_samples=40)  #make_wave() function is used for regression.
plt.scatter(X,y)
plt.legend(["regression"],loc=4)
plt.xlabel('feature')
plt.ylabel('target')
```
