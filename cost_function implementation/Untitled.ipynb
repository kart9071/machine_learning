```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(1)
x = np.random.rand(50, 1)
y = 2 + 3 * x + np.random.randn(50, 1)

# Plot the data
plt.scatter(x, y)
plt.xlabel('Independent variable')
plt.ylabel('Dependent variable')
plt.show()
def compute_cost(X, y, theta):
    """
    Compute the cost function for linear regression.
    
    Args:
    X -- feature matrix of shape (m, n+1), where m is the number of training examples and n is the number of features
    y -- target variable vector of shape (m, 1)
    theta -- parameter vector of shape (n+1, 1)
    
    Returns:
    J -- the cost value
    """
    m = len(y)
    predictions = X.dot(theta)
    square_errors = (predictions - y) ** 2
    J = 1/(2*m) * np.sum(square_errors)
    return J


# Add a column of ones to X for the intercept term
X = np.hstack((np.ones_like(x), x))

# Initialize theta with zeros
theta = np.zeros((2, 1))


# Compute the initial cost
initial_cost = compute_cost(X, y, theta)
print('Initial cost:', initial_cost)
```
