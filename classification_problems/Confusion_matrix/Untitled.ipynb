```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sb
import matplotlib.pyplot as plt

# example predicted and actual binary classification labels
y_pred = [1, 0, 1, 0, 1, 1, 0, 0, 1, 0]
y_true = [1, 1, 1, 0, 1, 0, 0, 1, 1, 0]

# create the confusion matrix
cm = confusion_matrix(y_true, y_pred)

# print the confusion matrix
print(cm)

sb.heatmap(cm,annot=True,cmap="Blues")
plt.xlabel("predicted")
plt.ylabel("Actual")
plt.title("Confusion matrix")
plt.show()
# here we can see that the whenever actual value is 0 the predicted false value will be 1 time so that the error will be 1.
#and whenever the actual value is 1 the predicted false value will be 2 times.
```
