```python
#logistic Regression

import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

data=pd.read_csv('exam_data.csv')
X=data[['StudyHours','Attendance']]
y=data['Passed']

plt.scatter(data['StudyHours'],data['Attendance'],c=data['Passed'])
plt.title('Result analysis')
plt.xlabel('Study Hours')
plt.ylabel('Attendance')
plt.show()

model=LogisticRegression()
model.fit(X,y)

predictions =model.predict([[5,90],[10,80]])
print(predictions)
```
